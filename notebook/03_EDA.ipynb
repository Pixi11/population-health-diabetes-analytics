```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(
    "../data/diabetes_cleaned.csv"
)
```

```python
df.head()
```

```python
# 1st Analysis
# Diabetes Distribution

df["Diabetes_012"].value_counts()
```

```python
plt.bar(
    df["Diabetes_012"].value_counts().sort_index().index,
    df["Diabetes_012"].value_counts().sort_index().values
    
)
plt.show()
```

```python
# 2nd Analysis
# Average BMI

df.groupby(
    "Diabetes_012"
)["BMI"].mean()
```

```python
sns.boxplot(
    x="Diabetes_012",
    y="BMI",
    data=df
)

plt.show()
```

```python
# 3rd Analysis
# High BP

pd.crosstab(
    df["Diabetes_012"],
    df["HighBP"]
)
```

```python
pd.crosstab(
    df["HighBP"],
    df["Diabetes_012"],
    normalize="index"
)*100
```

```python
# 3rd Analysis
# High Cholestrol

pd.crosstab(
    df["HighChol"],
    df["Diabetes_012"],
    normalize="index"
) * 100
```

```python
pd.crosstab(
    df["Smoker"],
    df["Diabetes_012"],
    normalize="index"
) * 100
```

```python
pd.crosstab(
    df["PhysActivity"],
    df["Diabetes_012"],
    normalize="index"
) * 100
```

```python
pd.crosstab(
    df["Stroke"],
    df["Diabetes_012"],
    normalize="index"
) * 100
```

```python
income_map={
     1:"<10K",
    2:"10K-15K",
    3:"15K-20K",
    4:"20K-25K",
    5:"25K-35K",
    6:"35K-50K",
    7:"50K-75K",
    8:"75K+"
}
df["Income_Level"]=df["Income"].map(income_map)
```

```python
pd.crosstab(
    df["Income_Level"],
    df["Diabetes_012"],
    normalize="index"
) * 100
```

```python
education_map = {
    1:"No School",
    2:"Elementary",
    3:"Some High School",
    4:"High School Graduate",
    5:"Some College",
    6:"College Graduate"
}

df["Education_Level"] = df["Education"].map(
    education_map
)
```

```python
pd.crosstab(
    df["Education_Level"],
    df["Diabetes_012"],
    normalize="index"
) * 100
```

```python
age_map = {
    1:"18-24",
    2:"25-29",
    3:"30-34",
    4:"35-39",
    5:"40-44",
    6:"45-49",
    7:"50-54",
    8:"55-59",
    9:"60-64",
    10:"65-69",
    11:"70-74",
    12:"75-79",
    13:"80+"
}

df["Age_Group"] = df["Age"].map(age_map)
```

```python
pd.crosstab(
    df["Age_Group"],
    df["Diabetes_012"],
    normalize="index"
) * 100
```

```python
df.to_csv(
    "../data/diabetes_featured.csv",
    index=False
)
```
