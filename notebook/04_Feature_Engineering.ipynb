```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
    "../data/diabetes_featured.csv"
)
df.head()
```

```python
df.info()
```

```python
# FEATURE 1 — BMI Category

def bmi_category(x):
    if x<18.5:
         return "Underweight"
    
    elif x < 25:
        return "Normal"

    elif x < 30:
        return "Overweight"

    else:
        return "Obese"   
    
```

```python
df["BMI_category"]=df["BMI"].apply(bmi_category)
```

```python
df["BMI_category"].value_counts()
```

```python
# FEATURE 2 — Risk Score

df["Risk_Score"] = (
      df["HighBP"]
    + df["HighChol"]
    + df["Smoker"]
    + df["Stroke"]
    + df["HeartDiseaseorAttack"]
)
```

```python
df["Risk_Score"].describe()
```

```python
# FEATURE 3 — Risk Level

def risk_level(x):

    if x <= 1:
        return "Low Risk"

    elif x <= 3:
        return "Medium Risk"

    else:
        return "High Risk"
```

```python
df["Risk_Level"] = (
    df["Risk_Score"]
    .apply(risk_level)
)
df["Risk_Level"].value_counts()
```

```python
# FEATURE 4 — Lifestyle Score

df["Lifestyle_Score"] = (

      df["PhysActivity"]

    + df["Fruits"]

    + df["Veggies"]

    - df["Smoker"]

)
df["Lifestyle_Score"].describe()
```

```python
# FEATURE 5 — Lifestyle Category

def lifestyle_category(x):

    if x <= 0:
        return "Poor"

    elif x <= 2:
        return "Average"

    else:
        return "Healthy"
```

```python
df["Lifestyle_Category"] = (
    df["Lifestyle_Score"]
    .apply(lifestyle_category)
)
df["Lifestyle_Category"]
```

```python
# FEATURE 6 — Healthcare Access Flag

df["Healthcare_Access_Risk"] = np.where(
    df["NoDocbcCost"] == 1,
    "Limited Access",
    "Adequate Access"
)
df["Healthcare_Access_Risk"]
```

```python
# FEATURE 7 — Health Burden Score

df["Health_Burden_Score"] = (

      df["GenHlth"]

    + df["MentHlth"]

    + df["PhysHlth"]

)
df["Health_Burden_Score"].describe()
```

```python
# FEATURE 8 — Health Burden Category

def burden_category(x):

    if x < 15:
        return "Low"

    elif x < 30:
        return "Medium"

    else:
        return "High"
```

```python
df["Health_Burden_Category"] = (
    df["Health_Burden_Score"]
    .apply(burden_category)
)
df["Health_Burden_Category"]
```
